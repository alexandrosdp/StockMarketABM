```python
def update_fundamental_value(time, fundamental_value, growth_rate, s):

    """
    Description
    -----------
    s : period of the business cycle
    Updates the fundamental value based on economic factors
    """

    # determine the cycle of the business growth
    i = time // (4 * s) + 1

    if time >= 4 * (i - 1) * s and time < (4 * i - 1) * s:
        return (growth_rate)
        # return self.growth_rate
    else:
        return (-growth_rate/2)
        # return -(self.growth_rate/2)
```

```python
import matplotlib.pyplot as plt
fundamental_value = 50
prices = [0]
for i in range(1000):
    prices.append(update_fundamental_value(i, prices[-1], 0.008, 25))

plt.plot(prices)    
```
